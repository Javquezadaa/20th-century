```python
# Step.5. Creating random data using pd.dataframe()

import pandas as pd

#DataFrame with numbers and text
data = {
    'numbers': [10, 20, 30, 40, 50],
    'fruits': ['apple', 'banana', 'cherry', 'date', 'elderberry']
}

df = pd.DataFrame(data)
df
```

```python
# Step. 6 Creating column using a lambda function
df['doubled'] = df['numbers'].apply(lambda x: x * 2)
df
```

```python
# Step 7: Creating a flag column 'is_citrus' based on whether the fruit is citrus or not
citrus_fruits = ['orange', 'lemon', 'lime', 'grapefruit', 'mandarin']
df['is_citrus'] = df['fruits'].apply(lambda x: x in citrus_fruits)

df
```
